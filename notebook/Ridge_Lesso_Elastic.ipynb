
Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

# Data Cleaning


```python
## Missing Values
dataset.isnull().sum() # Sends sum of missing val in each columns
dataset.isnull().any(axis=1) # Sends column in True or False\
dataset[dataset.isnull().any(axis=1)]
```

The Dataset is converted into two sets on Region from 122th index, we can make a new column based on region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122, "Region"] = 0      # 0 : Bejaia Region Dataset
dataset.loc[122:, "Region"] = 1      # 1 : Sidi-Bel Abbes Region Dataset
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
# Convert float to int
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]      # Again at 122 column we have same data columns heading in that
# So drop that
```

```python
df = df.drop(122).reset_index(drop = True)
# Removed 122 row as it is repeat of heading
df.iloc[[122]]
```

```python
df.columns
# Column names have extra SPACES
```

```python
## Fix spaces in Column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change the req column as integer data Datatype

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] =df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other col to float datatype



```python
objects = [features for features in df.columns if df[features].dtype == 'object']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()   # Here Classes Dtype wont change as if condition is present
```

```python
df.head()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index = False)
# Index wont b saved
```
